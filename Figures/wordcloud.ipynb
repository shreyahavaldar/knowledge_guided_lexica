```python
# Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
import random
from dlatk.lib import wordcloud
import rpy2.robjects as ro
from random import uniform as runif

import importlib
importlib.reload(wordcloud)
```

```python

def load_dict_from_file(filename):
    f = open(filename,'r')
    data=f.read()
    f.close()
    return eval(data)

def get_word_and_freq(filename, max_words=50):
    #read txt file
    loaded_dict = dict(load_dict_from_file(filename))
    keys_to_drop = ["responsibilities", "responsbilities"]
    for k in keys_to_drop:
        if k in loaded_dict:
            del loaded_dict[k]

    print(loaded_dict)
    words = list(loaded_dict.keys())
    freqs = [int(w * 100) for w in loaded_dict.values()]
    
    # keep only words with freq > 0
    words = [words[i] for i in range(len(words)) if freqs[i] > 0]
    freqs = [freqs[i] for i in range(len(freqs)) if freqs[i] > 0]
    
    # sort both lists by freqs
    words = sorted(words, key=lambda x: freqs[words.index(x)], reverse=True)
    words = [w for w in words if "0" not in str(w)]
    freqs = sorted(freqs, reverse=True)
    #pick max_words most frequent words
    words = words[:max_words]
    freqs = freqs[:max_words]
    print(words)
    return words, freqs

def _makeRandomColorTuples(num_rgb_tuples, rgb_bounds=[0, 0.39]):
    """Create a list of length num_rgb_tuples, where every element is a random hex color"""
    assert(num_rgb_tuples > 0)
    r = list(map(runif, [rgb_bounds[0]]*num_rgb_tuples, [rgb_bounds[1]]*num_rgb_tuples))
    g = list(map(runif, [rgb_bounds[0]]*num_rgb_tuples, [rgb_bounds[1]]*num_rgb_tuples))
    b = list(map(runif, [rgb_bounds[0]]*num_rgb_tuples, [rgb_bounds[1]]*num_rgb_tuples))
    ro_colors = list(map(ro.r.rgb, r, g, b))
    ro_color_list = ro.StrVector([x[0] for x in ro_colors])
    return ro_color_list

def _makeRed(num_rgb_tuples, rgb_bounds=[0.25, 0.85]):
    """Create a list of length num_rgb_tuples, where every element is a random hex color"""
    assert(num_rgb_tuples > 0)
    val = 0
    r = list(map(runif, [rgb_bounds[0]]*num_rgb_tuples, [rgb_bounds[1]]*num_rgb_tuples))
    g = list(map(runif, [rgb_bounds[0]*val]*num_rgb_tuples, [rgb_bounds[1]*val]*num_rgb_tuples))
    b = list(map(runif, [rgb_bounds[0]*val]*num_rgb_tuples, [rgb_bounds[1]*val]*num_rgb_tuples))
    ro_colors = list(map(ro.r.rgb, r, g, b))
    ro_color_list = ro.StrVector([x[0] for x in ro_colors])
    return ro_color_list

def _makeBlue(num_rgb_tuples, rgb_bounds=[0.15, 0.75]):
    """Create a list of length num_rgb_tuples, where every element is a random hex color"""
    assert(num_rgb_tuples > 0)
    val = 0
    r = list(map(runif, [rgb_bounds[0]*val]*num_rgb_tuples, [rgb_bounds[1]*val]*num_rgb_tuples))
    g = list(map(runif, [rgb_bounds[0]*val]*num_rgb_tuples, [rgb_bounds[1]*val]*num_rgb_tuples))
    b = list(map(runif, [rgb_bounds[0]]*num_rgb_tuples, [rgb_bounds[1]]*num_rgb_tuples))
    ro_colors = list(map(ro.r.rgb, r, g, b))
    ro_color_list = ro.StrVector([x[0] for x in ro_colors])
    return ro_color_list

def make_wordcloud(filename, color, output_prefix, max_words=100):

    words, freqs = get_word_and_freq(filename, max_words=max_words)
    
    if(color == "Blue"):
        color_list = _makeBlue(len(words))
    elif(color == "Red"):
        color_list = _makeRed(len(words))
    else:
        color_list = _makeRandomColorTuples(len(words))
    wordcloud.wordcloud(words, 
                    freqs, 
                    wordcloud_algorithm="amueller",
                    color_list=color_list,
                    output_prefix=output_prefix,
                    max_font_size=150,)

```

```python

make_wordcloud("lexica_coll.txt", "Red", "coll", max_words=250)
make_wordcloud("lexica_indv.txt", "Blue", "ind", max_words=200)
```
